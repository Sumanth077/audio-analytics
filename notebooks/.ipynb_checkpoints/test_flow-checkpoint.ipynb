## Start

### 1. Download audio from youtube video

```python
from pytube import YouTube
```

```python
yt = YouTube(youtube_url)
```

```python
audio_stream_itag = yt.streams.filter(only_audio=True)[0].itag
```

```python
stream = yt.streams.get_by_itag(audio_stream_itag)
```

```python
output_path = stream.download()
```

```python
output_path
```

```python
from moviepy.audio.io.AudioFileClip import AudioFileClip
```

```python
clip = AudioFileClip(output_path)
clip.write_audiofile(output_path[:-4] + ".mp3")
clip.close()
```

```python
import IPython
IPython.display.Audio(output_path[:-4] + ".mp3")
```
